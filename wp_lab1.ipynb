## 2. Typy danych
Komputer przechowuje wszystkie dane (liczby, teksty, instrukcje programu itp.) w swojej pamici operacyjnej (RAM) pod postaci cigw bitw.
Interpretacja danych: To, co oznacza cig bitw, zaley od typu danych i interpretacji przez program. Na przykad bajt 01000001 moe oznacza liczb 65 w kodzie dziesitnym, albo znak 'A' w kodzie ASCII  decyduje o tym kontekst. Dlatego w jzykach programowania wprowadzono pojcie typw danych, aby okreli, jak traktowa dane bitowe w pamici.
W jzyku C podstawowe typy to m.in.:
- **int**  liczba cakowita (integer). Zajmuje zazwyczaj 4 bajty (32 bity), co dla typu signed int daje zakres okoo od 2 147 483 648 do 2 147 483 647. Istniej take modyfikatory: short (zwykle 2 bajty), long (8 bajtw)  dajce odpowiednio mniejsze lub wiksze zakresy.
- **char**  typ znakowy (character). Zwykle przechowuje kod znaku ASCII. W praktyce uywamy char gwnie do przechowywania pojedynczych znakw.
- **float**  liczba zmiennoprzecinkowa pojedynczej precyzji, zajmuje 4 bajty, dokadno6-7 cyfr.
- **double**  liczba zmiennoprzecinkowa podwjnej precyzji. 8 bajtw, dokadno ok. 1516 cyfr znaczcych.
- **bool**  typ logiczny (ang. boolean) reprezentujcy warto prawda/fasz. Zmienna logiczna przyjmuje warto true (prawda) lub false (fasz). W rzeczywistoci przechowywana jest zazwyczaj w 1 bicie (gdzie 0 oznacza false, a 1 oznacza true).

## 3. Zmienne

Zmienne to podstawowe pojcie w programowaniu  oznaczaj nazwane miejsca w pamici, w ktrych moemy przechowa dane okrelonego typu. Kada zmienna ma:
- typ,
 - nazw nadan przez programist (powinny zaczyna si liter/underscore, mog zawiera cyfry, nie uywamy spacji ani polskich znakw). Dobrze, by nazwy byy czytelne i opisowe  np. suma, indeks, pozycjaX zamiast nic nie mwicych a, b, c.
- warto (ktra moe si zmienia podczas dziaania programu).

**Deklaracja i inicjalizacja**: W jzyku C zanim uyjemy zmiennej, musimy j zadeklarowa, podajc typ i nazw, ewentualnie z nadaniem pocztkowej wartoci (inicjalizacj).





## Zad 1.

```python
#include <stdio.h>
#include <stdbool.h>

int main() {
    int rok_urodzenia = 2005;
    float wzrost = 3.14f;
    char znak = 'A';
    bool prawko = true;

    printf("Zmienna 'rok_urodzenia' ma warto: %d i zajmuje %zu bajty\n", rok_urodzenia, sizeof(rok_urodzenia));
    printf("Zmienna 'wzrost' ma warto: %.2f i zajmuje %zu bajty\n", wzrost, sizeof(wzrost));
    printf("Zmienna 'znak' ma warto: %c i zajmuje %zu bajty\n", znak, sizeof(znak));
    printf("Zmienna 'prawko' ma warto: %d (true=1, false=0) i zajmuje %zu bajt\n", prawko, sizeof(prawko));

    return 0;
}
```

### 3.1 Zmienne lokalne i globalne

**Zmienne globalne** (deklarowane poza funkcjami)  te istniej przez cay czas dziaania programu.

**Zmienne lokalne** - tworzone wewntrz funkcji, istniej tylko podczas wykonywania danej funkcji (po jej zakoczeniu pami moe by uyta ponownie do innych celw). Ich zakres wanoci ograniczony jest do bloku { ... }, w ktrym zostay zadeklarowane.

## Zad 2.

```python
#include <stdio.h>

// Zmienna globalna  istnieje przez cay czas dziaania programu
int licznik = 0;

void funkcja() {
    // Zmienna lokalna  tworzona na nowo przy kadym wywoaniu funkcji
    int licznik_funkcji = 5;
    printf("W funkcji: licznik_funkcji = %d\n", licznik_funkcji);
}

int main() {
    printf("W main (globalna): licznik = %d\n", licznik);
    funkcja();
    printf("Po powrocie do main: licznik = %d\n", licznik);
    return 0;
}
```

## 4. Operatory

**Operatory** to symbole oznaczajce wykonywanie pewnych operacji na danych:
- **Operatory przypisania**: = suy do przypisania wartoci do zmiennej. Po lewej stronie musi by zmienna, po prawej wyraenie. Np. x = 5, y = a + 3. Istniej te skrcone formy jak +=, -=, *=, /=, %= ktre cz operacj z przypisaniem (np. x += 5; oznacza x = x + 5;).
- **Operatory porwnania** (relacyjne): su do porwnywania wartoci, zwracaj wynik typu logicznego true/false. Mamy:== (rwno porwnania, uwaga: dwa znaki = odrniaj od pojedynczego = przypisania),!= (rne),< (mniejsze), > (wiksze), <= (mniejsze lub rwne), >= (wiksze lub rwne).

- **Operatory arytmetyczne**: su do wykonywania podstawowych dziaa matematycznych:+ (dodawanie), - (odejmowanie), * (mnoenie), / (dzielenie), % (modulo  reszta z dzielenia).Przykady: z = x + y;, a = b * 2;, c = c - 1;.

Uwaga: operator % jest stosowalny tylko do liczb cakowitych (np. 7 % 3 to 1).

 Dzielenie / dwch liczb cakowitych zwraca cz cakowit wyniku (np. 7 / 3 to 2 w C). Jeli chcemy uzyska wynik rzeczywisty, przynajmniej jeden operand musi by typu zmiennoprzecinkowego (np. 7.0 / 3 to  2.3333).
- **Operatory inkrementacji/dekrementacji**: ++ zwiksza warto zmiennej o 1, a -- zmniejsza o 1. Mona ich uywa w formie prefiksowej (++i, --i) lub postfiksowej (i++, i--), ktre rni si momentem zwrcenia wartoci.
